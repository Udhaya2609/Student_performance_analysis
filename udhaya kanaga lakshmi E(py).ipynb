```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#load data
df = pd.read_csv(r"C:\Users\acer\Desktop\NM PROJECT\student_data.csv")

print("Data successfully loaded from CSV!")
display(df.head())
```

```python
#check column and data type
df.info()
```

```python
#get statistical summary
df.describe()
```

```python
#check null values
print(df.isnull().sum())
```

```python
# Gender-wise average score
df.groupby('gender')['average score'].mean()

# Race/Ethnicity group-wise average score
df.groupby('race/ethnicity')['average score'].mean()

# Parental education-wise average score
df.groupby('parental level of education')['average score'].mean()

# Test preparation course-wise average score
df.groupby('test preparation course')['average score'].mean()

# Lunch type-wise average score
df.groupby('lunch')['average score'].mean()

# Result (pass/fail)-wise total score
df.groupby('result')['total score'].mean()
```

```python
# Only numeric columns use  - df.corr()
df[['math score', 'reading score', 'writing score', 
    'total score', 'average score']].corr()
```

```python
#visualization
#(gender distribution)
plt.figure(figsize=(7,7))
df['gender'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], shadow=True, startangle=140)
plt.title('Gender Distribution')
plt.show()
```

```python
#parental education vs math score
plt.figure(figsize=(12,6))
sns.barplot(x='parental level of education', y='math score', data=df, palette='viridis')
plt.xticks(rotation=45)
plt.title('Impact of Parental Education on Math Scores')
plt.show()
```

```python
#correlation headmap
plt.figure(figsize=(8,6))
# Selecting only numeric columns for correlation
numeric_df = df[['math score', 'reading score', 'writing score']]
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Scores')
plt.show()
```

```python
# Distribution of math scores
df['math score'].plot(kind='hist',
                       title='Distribution of Math Scores',
                       bins=20,
                       figsize=(8, 5),
                       color='coral',
                       edgecolor='black')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Average score by index (student performance trend)
df['average score'].plot(title='Student Average Score Trend', figsize=(10, 4))
plt.xlabel('Student Index')
plt.ylabel('Average Score')
plt.tight_layout()
plt.show()
```
